# Partie 2 (Clustering )

1. Essayer de construire le modèle de clustering en utilisant sklearn ainsi algorithme Kmeans,

```python
from sklearn.cluster import KMeans
```

```python
#Pour la visualisation, on utilise juste les variables BALANCE_FREQUENCY et PURCHASES
X = df.iloc[:, [2, 3]].values
```

```python
#Ou bien Appliquons la PCA pour la réduction de dimension
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df)
pca_data = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_data.head()
```

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)       
    wcss.append(kmeans.inertia_)    # inertia method returns wcss for that model
```

2. Définir le K nécessaire pour avoir un modèle efficace selon la méthode de Elbow.

```python
import seaborn as sns
sns.lineplot(range(1, 10), wcss,marker='o',color='red') 
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

la meilleur valeur dr K est 4 --> 4 clusters

3. Présenter les clusters obtenues dans un graphe en utilisant matplotlib.

```python
kmeans = KMeans(n_clusters=4 , init = 'k-means++', random_state = 42)
pca_preds = kmeans.fit_predict(pca_data)  #Method 1 : using PCA algorithm
x_preds = kmeans.fit_predict(X)         #Method 2 : using 2 columns from our dataset
```

```python
# Visualisation des 4 clusters
plt.figure(figsize=(10,7))
# plot data
sns.scatterplot(X[x_preds == 0, 0], X[x_preds == 0, 1], color = 'purple', label ='Cluster 1',s=50)
sns.scatterplot(X[x_preds == 1, 0], X[x_preds == 1, 1], color = 'yellow', label = 'Cluster 2',s=50)
sns.scatterplot(X[x_preds == 2, 0], X[x_preds == 2, 1], color = 'blue', label ='Cluster 3',s=50)
sns.scatterplot(X[x_preds == 3, 0], X[x_preds == 3, 1], color = 'green', label = 'Cluster 4',s=50)
# plot centroids
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],color = 'red',label = 'Centroids',s=200,marker='^')

plt.grid(False)
plt.title('Segmentation Client')
plt.xlabel('BALANCE_FREQUENCY')
plt.ylabel('PURCHASESX ')
plt.legend()
plt.show()
```

5. Refaire la même chose en utilisant l’algorithme fuzzy cmeans « il faut utiliser la bibliothèque
skfuzzy ».

```python
#!pip install -U scikit-fuzzy
import skfuzzy as fuzz
import numpy as np
```

```python
colors = ['blue', 'orange', 'green', 'red', 'c', 'm', 'yellow', 'k', 'pink', 'ForestGreen']
# Importation du dataset
data = df.values
data = data[:, 1:4]

xpts = data[:, 0]
ypts = data[:, 1]

fig1, axes1 = plt.subplots(3, 3, figsize=(10, 10))
#Pour empiler les points(x,y)
alldata = np.vstack((xpts, ypts))
#Initialisation du tableau des fuzzy clusters
fpcs = []

for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, \
    fpc = fuzz.cluster.cmeans(alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)
    
    #On ajoute à chaque fois au tableau le fuzzy cluster qu'on a
    fpcs.append(fpc)
    
    # Pour traces les clusters 
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_membership == j],
        ypts[cluster_membership == j], '.', color=colors[j])
    
    # Pour marquer le centre de chaque cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')
    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc), size=12)
    ax.axis('off')
    
# Pour ajuster l'emplacement des axes
fig1.tight_layout()
#Pour dessiner plusieurs tracés sur une seule figure
fig2, ax2 = plt.subplots()
ax2.plot(np.r_[2:11], fpcs, color='r')
ax2.set_title("How Number of Clusters Change FPC?")
ax2.set_xlabel("Number of centers")
ax2.set_ylabel("Fuzzy partition coefficient")
plt.show()
```

6. Refaire la même chose en utilisant l’algorithme DBSCAN et Tsne.

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
#features
ft = df.iloc[:, 0:17].values
```

```python
# Scale the data
scaler = StandardScaler()
ft = scaler.fit_transform(ft)
```

```python
# Apply DBSCAN algorithm
dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan.fit(ft)
```

```python
# Get the labels and the number of clusters
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
# Print the number of clusters and the labels
print("Number of clusters: ", n_clusters)
print("Labels: ", labels)
```

```python
# Visualize the clusters
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'orange', 'purple', 'brown','pink','purple','gray']
for i in range(-1, n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], s=50, c=colors[i % len(colors)], label='Cluster {}'.format(i))
plt.title('DBSCAN Clustering')

plt.legend()
plt.show()
```

```python
# Perform t-SNE to reduce the dimensionality of the dataset to 2
# Perform t-SNE clustering
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200)
ft_tsne = tsne.fit_transform(ft)
```

```python
# Apply DBSCAN algorithm on the 2-dimensional dataset
dbscan = DBSCAN(eps=4, min_samples=5)
dbscan.fit(ft_tsne)
```

```python
# Get the labels and the number of clusters
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
# Visualize the clusters
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'brown', 'gray','pink','purple' ]
for i in range(n_clusters):
    plt.scatter(ft_tsne[labels == i, 0], ft_tsne[labels == i, 1], s=50, c=colors[i % len(colors)], label=f'Cluster {i+1}')
plt.scatter(ft_tsne[labels == -1, 0], ft_tsne[labels == -1, 1], s=50, c='k', label='Noise')
plt.legend()
plt.show()
```
